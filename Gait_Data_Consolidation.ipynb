```python
# function that combines all logs in a folder
def file_combine(folder):
    combined_csv = pd.DataFrame()
    
    #iterates through all log files in a headset
    for f in all_filenames:
        file = pd.read_csv(f)
        print(f)

        #add and edit columns for dates, times, and IDs
        file.insert(0, 'Participant ID', str(folder))
        file.insert(0, 'Info Group', str(f))
        combined_csv = combined_csv.append(file)
        print(combined_csv.shape)
    
    combined_csv.head()
    return(combined_csv)
```

```python
#utilize file aggregator and iterate through all folders to create a single large CSV
full_csv = pd.DataFrame()
extension = "csv"
#iterate through all folders
for folder in all_folders:
    #uncomment below for debugging
    #print(folder)
    
    #move to file for aggregation
    os.chdir(str(folder))
    print(str(folder))
    #iterate through all files in folder
    all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
    
    #uncomment below for debugging
    #print(all_filenames)
    
    #create a combinedfile called "csv"
    csv = file_combine(folder)
    #add the new combined file to the master combined file
    full_csv = full_csv.append(csv)
    #go back to the main directory
    os.chdir("..")
```

### Variance Testing

```python
# One-Way ANOVA Test for stride length across experimentation modes.
f_oneway(df_comb_dt_con['stride_lengths'], df_comb_dt_fat['stride_lengths'], df_comb_st_con['stride_lengths'], df_comb_st_fat['stride_lengths'])
```

##### There exist a statistically significant difference between stride lenghts across experimentation modes.

```python
# One-Way ANOVA Test for stride times across experimentation modes.
f_oneway(df_comb_dt_con['stride_times'], df_comb_dt_fat['stride_times'], df_comb_st_con['stride_times'], df_comb_st_fat['stride_times'])
```
